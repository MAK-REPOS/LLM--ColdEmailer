```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader('https://www.google.com/about/careers/applications/jobs/results/')
page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate
prompt_extract = PromptTemplate.from_template('''###scraped text from website :
{page_data}
###instruction:
the scraped text is from the careers page of a website.
your job is to extract the job postings and return them in JSON format container.
following keys :'role' , 'experience' , 'skills' and 'description'.
###only return valid JSON (NO PREAMBLE):
'''
)

chain_extract = prompt_extract | llm
res = chain_extract.invoke(input = {'page_data' : page_data})
print(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser
json_parser = JsonOutputParser()
json_res = json_parser.parse(res.content)
json_res
```

```python
import pandas as pd
df = pd.read_csv(r'my_portfolio.csv')
df
```

```python
import uuid
import chromadb
client = chromadb.PersistentClient('vectorstore')
collection = client.get_or_create_collection(name = 'portfolio')
if not collection.count():
    for _ , row in df.iterrows():
        collection.add(documents = row['Techstack'] , 
                       metadatas = {'links' : row['Links']},
                       ids = [str(uuid.uuid4())])
```

```python
links = collection.query(query_texts=['experience in python','expertise in react native '],n_results = 2).get('metadatas' , [])
links
```
